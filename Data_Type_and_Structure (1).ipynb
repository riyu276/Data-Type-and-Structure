Question 19: Why are dictionaries considered unordered, and how does that afect data retrieval?

Answer :- Historically (prior to python 3.6), dictionaries were unordered because:-

* They used hash tables fpr efficiency.
* Keys were stored based on their hash   value,not the order of insertion.
* There was no guarantee that iterating over a dictionary return items in the order you added them.

How this affects data retrieval:-

* dict[key] => constant -time lookup by key (0(1))
* order-based indexing => not supported (dict[0] is invalid)
* iteration order => requires conversion(e.g., sorted(dict.items()))
